```python
import csv
from operator import itemgetter
from tkinter import *

def get_mpg(data, make, model, trans, fuel):
    for row in data:
        if row['make'] == make and row['model'] == model and row['transmission'] == trans and row['fuel_type'] == fuel:
            return row['mpg']
    return "Not Found"

def on_select(event):
    make = make_var.get()
    model = model_var.get()
    trans = trans_var.get()
    fuel = fuel_var.get()
    mpg = get_mpg(data, make, model, trans, fuel)
    mpg_label.config(text=mpg)

#Read CSV file and select specific columns
with open('vehicles.csv', 'r') as f:
    reader = csv.DictReader(f)
    data = [{'make': row['make'], 'model': row['model'], 'transmission': row['transmission'], 'fuel_type': row['fuel_type'], 'mpg': row['mpg']} for row in reader]
    data = sorted(data, key=itemgetter('make', 'model'))

# Create a window
root = Tk()
root.title("Vehicles")

# Create a variable for the dropdown
make_var = StringVar(root)
make_var.set("Make") # default value

# Create a dropdown for make
make_dropdown = OptionMenu(root, make_var, *set(row["make"] for row in data))
make_dropdown.pack()

# Create a variable for the dropdown
model_var = StringVar(root)
model_var.set("Model") # default value

# Create a dropdown for model
model_dropdown = OptionMenu(root, model_var, *set(row["model"] for row in data))
model_dropdown.pack()

# Create a variable for the dropdown
trans_var = StringVar(root)
trans_var.set("Transmission") # default value

# Create a dropdown for transmission
trans_dropdown = OptionMenu(root, trans_var, *set(row["transmission"] for row in data))
trans_dropdown.pack()

# Create a variable for the dropdown
fuel_var = StringVar(root)
fuel_var.set("Fuel Type") # default value

# Create a dropdown for fuel type
fuel_dropdown = OptionMenu(root, fuel_var, *set(row["fuel_type"] for row in data))
fuel_dropdown.pack()

# Create a label to display mpg
mpg_label = Label(root, text="")
mpg_label.pack()

# Add command to dropdown
make_dropdown.bind("<<OptionMenuSelect>>", on_select)
model_dropdown.bind("<<OptionMenuSelect>>", on_select)
trans_dropdown.bind("<<OptionMenuSelect>>", on_select)
fuel_dropdown.bind("<<OptionMenuSelect>>
```
